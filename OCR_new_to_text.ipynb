```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import pytesseract
import cv2
```

```python
img = cv2.imread("1.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# blur
blur = cv2.GaussianBlur(gray, (0,0), sigmaX=33, sigmaY=33)
# divide
divide = cv2.divide(gray, blur, scale=255)

# otsu threshold
thresh = cv2.threshold(divide, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]

# apply morphology
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3,3))
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
image_up = cv2.pyrUp(divide)
image_up = cv2.pyrUp(image_up)

blur = cv2.GaussianBlur(image_up, (0,0), sigmaX=33, sigmaY=33)
divide = cv2.divide(image_up, blur, scale=255)
image_up = cv2.pyrUp(divide)
image_up = cv2.pyrUp(image_up)

plt.imshow(cv2.cvtColor(image_up, cv2.COLOR_BGR2RGB))
plt.title('my picture')
plt.show()

cv2.imwrite("2.png",image_up)
```

```python
pytesseract.pytesseract.tesseract_cmd = r"E:\OCR\tesseract.exe"
text_1=(pytesseract.image_to_string(image_up, lang ="eng").lower())
print("---------------------------------------------------------------------")
print("Text: ",text_1)
```
